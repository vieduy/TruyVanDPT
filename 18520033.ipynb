```
import pandas as pd
import requests
```

```
import json

# Get headlines from tinhte.vn
# I got that link when i try to click button 'read more' and check the 'network' tab in the inspect of browser, then i saw that browser requested to that link.
# Warning: because the token can expire so you must get url one time in everytime get headlines. 
url = 'https://tinhte.vn/appforo/index.php?lists/5/threads&_bdImageApiThreadThumbnailHeight=193&_bdImageApiThreadThumbnailWidth=312&fields_include=list_item%2Ccontent.thread_id%2Ccontent.thread_title%2Ccontent.first_post.post_body_plain_text%2Ccontent.first_post.post_id%2Ccontent.first_post.post_is_liked%2Ccontent.first_post.post_like_count%2Ccontent.first_post.poster_rank%2Ccontent.first_post.poster_has_verified_badge%2Ccontent.first_post.post_body%2Ccontent.first_post.attachments%2Ccontent.first_post.post_body_html%2Ccontent.thread_image.link%2Ccontent.thread_thumbnail.link%2Ccontent.links.permalink%2Ccontent.links.posts%2Ccontent.thread_create_date%2Ccontent.creator_user_id%2Ccontent.creator_username%2Ccontent.creator_has_verified_badge%2Ccontent.links.first_poster_avatar%2Ccontent.thread_post_count%2Ccontent.thread_view_count%2Ccontent.forum.forum_id%2Ccontent.forum.forum_title%2Ccontent.forum.links.permalink%2Ccontent.thread_tags%2Ccontent.permissions%2Ccontent.user_is_ignored%2Ccontent.latest_posts&limit=200&page=1&oauth_token=0%2C1600510733%2Ca836b39cb73d4cfb6f54bf0f66ebf2c2%2Clxi7g2zolu' # các bạn thay link của trang mình cần lấy dữ liệu tại đây
headers = {'Content-Type': 'application/json'}

def get_page_content(url):
   # Get response from request to that url 
   page = requests.get(url, headers=headers)

   # Load json file after got response string
   text = json.loads(page.text)
   return text['threads']

soup = get_page_content(url)
print(soup)
```

```
from datetime import datetime

headlines = []
date = []
for i in range(0, len(soup)):
  # Get headlines if thread_create_date within 1 month from now
  if (soup[i]['thread_create_date'] > int(datetime.now().strftime("%s")) - 3600*24*30):
    headlines.append(soup[i]['thread_title'])

    # Convert date type
    date.append(datetime.fromtimestamp(soup[i]['thread_create_date']).strftime("%A, %B %d, %Y %I:%M:%S"))

print(headlines)
print(date)
```

```
from IPython.display import display, HTML

# Create as a dictionary
dict = {'Headlines': headlines, 'Date': date} 
df = pd.DataFrame(dict) 

# saving the dataframe 
df.to_csv('data.csv')

# display headlines with date create
display(HTML(df.to_html()))
```
